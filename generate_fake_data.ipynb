```python
photo = pd.read_csv("YN_LDGS_PROPRERTY_PHOTO_URL_202312.csv")[["LDGS_CD", "GSRM_CD", "GSRM_IMAGE_URL"]].rename(columns={"LDGS_CD": "stay_seq", "GSRM_CD": "room_seq", "GSRM_IMAGE_URL": "room_image_url"})
photo
```

```python
photo["image_seq"] = photo.groupby(["stay_seq", "room_seq"]).cumcount() + 1
photo
```

```python
room_type = pd.read_csv("YN_LDGS_ROOMTYPE_DC_202312.csv")\
.rename(columns={"LDGS_CD": "stay_seq", "LDGS_NM": "stay_name", "GSRM_CD": "room_seq", "GSRM_NM": "room_name", "GSRM_IMAGE_URL": "room_image_url", "GSRM_DC": "room_discription"})
room_type
```

```python
stay_type = pd.read_csv("YN_LDGS_STAYNG_TY_202312.csv").rename(columns={"LDGS_CD": "stay_seq", "LDGS_NM": "stay_name", "LDGS_TY_CD": "stay_type", "LDGMNT_TY_NM": "stay_type_name"})[["stay_seq", "stay_type", "stay_type_name"]]
stay_type
```

```python
join_stay = pd.merge(room_type, stay_type, on=["stay_seq"], how="inner")
stay = pd.merge(join_stay, photo, on=["stay_seq", "room_seq"], how="inner")
stay
```

```python
stay["stay_type_name"].info()
```

```python
stay["stay_type_name"].value_counts()
```

```python
stay["stay_type_name"] = stay["stay_type_name"].str.replace("Pension", "펜션").str.replace("Motel", "모텔").str.replace("Hotel", "호텔").str.replace("GuestHouse", "게스트하우스")
stay["stay_type_name"]
```

```python
stay
```

```python
mask = (
    (stay["stay_name"].str.contains("펜션") & ~stay["stay_type_name"].str.contains("펜션")) |
    (stay["stay_name"].str.contains("호텔") & ~stay["stay_type_name"].str.contains("호텔")) |
    (stay["stay_name"].str.contains("게스트하우스") & ~stay["stay_type_name"].str.contains("게스트하우스")) |
    (stay["stay_name"].str.contains("모텔") & ~stay["stay_type_name"].str.contains("모텔")) |
    (~stay["stay_name"].str.contains("펜션") & stay["stay_type_name"].str.contains("펜션")) |
    (~stay["stay_name"].str.contains("호텔") & stay["stay_type_name"].str.contains("호텔")) |
    (~stay["stay_name"].str.contains("게스트하우스") & stay["stay_type_name"].str.contains("게스트하우스")) |
    (~stay["stay_name"].str.contains("모텔") & stay["stay_type_name"].str.contains("모텔"))
)
mask
```

```python
stay[~mask]
```

```python
filterd_stay = stay[~mask].reset_index(drop=True)
filterd_stay
```

```python
filterd_stay = filterd_stay.groupby(["stay_seq", "room_seq"]).filter(lambda x: len(x) >= 3).sort_values(["stay_seq", "room_seq"]).reset_index(drop=True)
filterd_stay
```
