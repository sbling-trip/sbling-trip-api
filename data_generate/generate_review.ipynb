```python
def combine_reviews(reviews_list, max_combinations=2):
    # 무작위로 1개에서 max_combinations 개의 리뷰를 조합합니다.
    num_reviews_to_combine = random.randint(1, max_combinations)
    selected_reviews = random.choices(reviews_list, k=num_reviews_to_combine)
    combined_review = " ".join(selected_reviews)
    return combined_review


def generate_image_urls(n):
    result = []
    for _ in range(n):
        if random.randint(0, 100) > 10:
            result.append(random.choice(room_image_list))
        else:
            result.append(fake.image_url())
    return result


def generate_korean_reviews(n):
    reviews = []

    for _ in range(n):
        num_images = random.randint(0, 5)  # 이미지 URL 개수 무작위 선택
        review_score = random.randint(1, 5)  # 리뷰 점수 무작위 선택
        if review_score < 2: # 낮은 점수 다시 돌리기
            review_score = random.randint(1, 5)  # 리뷰 점수 무작위 선택
            
        
        # 리뷰 점수에 따라 제목과 내용 조합
        if review_score <= 2:  # 부정적인 리뷰
            review_title = combine_reviews(bad_review_titles, max_combinations=2)
            review_content = combine_reviews(bad_review_contents, max_combinations=3)
        elif review_score == 3: # 그저 그런 리뷰
            review_title = random.choice(jeju_neutral_review_titles)
            review_content = random.choice(jeju_neutral_review_contents)
        else:  # 긍정적인 리뷰
            review_title = combine_reviews(review_titles, max_combinations=2)
            review_content = combine_reviews(review_contents, max_combinations=3)

        review = {
            "user_seq": random.randint(4, 1000),
            # "stay_seq": random.randint(1, 500),
            "room_seq": random.randint(1, 508),
            "review_title": review_title,
            "review_content": review_content,
            "review_score": review_score,
            "review_image_url_list": generate_image_urls(num_images),
            "created_at": fake.date_time_between(datetime.date(2023, 1, 1))
            # "modified_at": fake.date_time_this_year()
        }
        reviews.append(review)

    return pd.DataFrame(reviews)
```
