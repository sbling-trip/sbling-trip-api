```python
def generate_fake_reservation():
    check_in_date = fake.date_between(start_date='-1M', end_date='+3M')
    check_out_date = check_in_date + timedelta(days=random.randint(1, 30))  # 체크아웃은 체크인 후 1~30일 이내
    booking_date = fake.date_time_between(start_date='-1y', end_date=check_in_date)
    payment_status = random.choice(['paid', 'unpaid'])
    
    return {
        'room_seq': random.randint(1, 508),
        'user_seq': random.randint(1, 508),
        'check_in_date': check_in_date,
        'check_out_date': check_out_date,
        'adult_guest_count': random.randint(1, 4),
        'child_guest_count': random.randint(0, 3) if random.choice([True, False]) else 0,
        'reservation_status': random.choice(['confirmed', 'cancelled']) if payment_status == "paid" else random.choice(['pending', 'cancelled']),
        'booking_date': booking_date,
        'payment_status': payment_status,
        'special_requests': random.choice(special_request_list) if random.choice([True, False]) else None
    }
```

```python
fake_reservations = [generate_fake_reservation() for _ in range(10000)]
df_reservations = pd.DataFrame(fake_reservations)
print(df_reservations.head())
```

```python
room = pd.read_csv("room_info_v2.csv", index_col=0).reset_index(drop=True)[["room_seq", "stay_seq", "min_people", "max_people"]]
room
```

```python
result = pd.merge(df_reservations, room, on=["room_seq"], how="inner")
result
```

```python
# 최소 최대 인원 조건에 맞지 않을 경우 해결
def adjust_guest_counts(row):
    min_people = row['min_people']
    max_people = row['max_people']
    total_guests = row['adult_guest_count'] + row['child_guest_count']
    if total_guests < min_people:
        row['adult_guest_count'] = min_people - row['child_guest_count']
    elif total_guests > max_people:
        row['child_guest_count'] = max_people - row['adult_guest_count']
    return row

# Apply the adjustments across the DataFrame
result = result.apply(adjust_guest_counts, axis=1)
```
