```python
import random
def assign_stay_type(stay_name):
    if '모텔' in stay_name:
        return 1
    elif '호텔' in stay_name:
        return 2
    elif '펜션' in stay_name:
        return 3
    elif '게스트' in stay_name:
        return 4
    elif '텔' in stay_name:
        return random.choice([1, 2])
    else:
        return random.choice([1, 2, 3, 4])

stay_info['stay_type'] = stay_info['stay_name'].apply(assign_stay_type)
stay_info
```

```python
motel_image = motel.groupby('room_seq').agg({
    'room_name': 'first',  # 첫 번째 room_name을 사용
    'room_description': 'first',  # 첫 번째 room_description을 사용
    'room_image_url': lambda x: list(x)  # 모든 room_image_url을 리스트로 모음
}).reset_index()

hotel_image = hotel.groupby('room_seq').agg({
    'room_name': 'first',  # 첫 번째 room_name을 사용
    'room_description': 'first',  # 첫 번째 room_description을 사용
    'room_image_url': lambda x: list(x)  # 모든 room_image_url을 리스트로 모음
}).reset_index()

guest_image = guest_house.groupby('room_seq').agg({
    'room_name': 'first',  # 첫 번째 room_name을 사용
    'room_description': 'first',  # 첫 번째 room_description을 사용
    'room_image_url': lambda x: list(x)  # 모든 room_image_url을 리스트로 모음
}).reset_index()

pension_image = pension.groupby('room_seq').agg({
    'room_name': 'first',  # 첫 번째 room_name을 사용
    'room_description': 'first',  # 첫 번째 room_description을 사용
    'room_image_url': lambda x: list(x)  # 모든 room_image_url을 리스트로 모음
}).reset_index()
```

```python
motel_room_list = [("일반실", 45000, 55000), ("준특실", 56000, 60000), ("특실", 61000, 70000), ("VIP실", 71000, 80000), ("VVIP실", 81000, 100000)]
hotel_room_list = [("스탠다드", 55000, 65000), ("트윈", 66000, 75000), ("스위트", 76000, 85000), ("디럭스", 86000, 100000), ("프리미엄", 120000, 150000)]
pension_room_list = [("일반실", 70000, 80000), ("가족실", 100000, 120000)]
guest_room_list = [("도미토리 남성", 35000, 55000), ("도미토리 여성", 56000, 65000), ("단체룸", 65000, 75000)]

def random_generate_room(stay_type):
    result = []

    if stay_type == 1:
        room_count = random.randrange(1, 6)
        seleted_list = random.sample(motel_room_list, room_count)
    elif stay_type == 2:
        room_count = random.randrange(1, 6)
        seleted_list = random.sample(hotel_room_list, room_count)
    elif stay_type == 3:
        room_count = random.randrange(1, 3)
        seleted_list = random.sample(pension_room_list, room_count)
    else:
        room_count = random.randrange(1, 4)
        seleted_list = random.sample(guest_room_list, room_count)
    
    for room_type, min_money, max_money in seleted_list:
        random_int = random.randint(min_money, max_money)
        modified_int = (random_int // 1000) * 1000
        
        result.append((room_type, modified_int))
    result.sort(key=lambda x: x[1])
    return result    
```

```python
ott_list = ["넷플릭스", "웨이브", "애플TV", "디즈니", "티빙"]
toilet_list = ["스파", "욕조", "비데"]
room_option_list = ["공기청정기", "스타일러", "고사양 게이밍PC", "안마의자", "스마트 TV", "빔 프로젝트", "노래방 기기"]
special_room_list = ["금연룸", "이벤트룸", "조식 배달 서비스"]

global_stay_seq = 0
global_ott_set = set()
global_toilet_set = set()
global_room_option_set = set()
global_special_room_list = set()

def random_generate_room_option(stay_seq):
    global global_stay_seq
    global global_ott_set
    global global_toilet_set
    global global_room_option_set
    global global_special_room_list

    ott_count = random.randrange(1, 3)
    toilet_count = random.randrange(1, 2)
    room_option_count = random.randrange(1, 4)
    special_room_count = random.randrange(1, 2)

    ott_select = set(random.sample(ott_list, ott_count))
    toilet_select = set(random.sample(toilet_list, toilet_count))
    room_select = set(random.sample(room_option_list, room_option_count))
    special_room_select = set(random.sample(special_room_list, special_room_count))

    if global_stay_seq == stay_seq:
        global_ott_set = global_ott_set | ott_select
        global_toilet_set = global_toilet_set | toilet_select
        global_room_option_set = global_room_option_set | room_select
        global_special_room_list = global_special_room_list | special_room_select
    else:
        global_stay_seq = stay_seq
        global_ott_set = ott_select
        global_toilet_set = toilet_select
        global_room_option_set = room_select
        global_special_room_list = special_room_select

    a = sorted(list(global_ott_set))
    b = sorted(list(global_toilet_set))
    c = sorted(list(global_room_option_set))
    d = sorted(list(global_special_room_list))

    return pd.Series([a, b, c, d])
    
```

```python
stay_info[["ott_service", "toilet_option", "room_option", "special_room_option"]] = stay_info["stay_seq"].apply(random_generate_room_option)
stay_info
```
